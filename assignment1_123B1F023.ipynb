```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#drawing a simple line
img = np.zeros((400,400,3),dtype = 'uint8')
cv2.line(img,(50,50),(350,350),(150,0,180),10)
cv2.line(img,(100,100),(300,300),(150,0,180),10)
cv2_imshow(img)
```

```python
#drawing a rectangle alnog with the previous line
cv2.rectangle(img,(100,100),(300,300),(100,0,100),10)
cv2_imshow(img)
#drawinh a reactangle without using previous imaage that included a line
img1 = np.zeros((400,400,3),dtype = 'uint8')
cv2.rectangle(img1,(100,100),(300,300),(100,0,180),10)
cv2_imshow(img1)
```

```python
#drawing a hollow circle
img2 = np.zeros((400,400,3),dtype = 'uint8')
cv2.circle(img2,(200,200),100,(100,0,100),10)
cv2_imshow(img2)
```

```python
#drawing a triangle
img3 = np.zeros((400,400,3),dtype = 'uint8')

#defining 2 tuples that represenmt 3 vertices of the triangle
p1 = (300,300)
p2 = (100,200)
p3 = (180,150)

#defines the 3 sides of the triangle by using cv2.line function thrice
cv2.line(img3,p1,p2,(100,50,100),8)
cv2.line(img3,p2,p3,(100,50,100),8)
cv2.line(img3,p3,p1,(100,50,100),8)
cv2_imshow(img3)
```
